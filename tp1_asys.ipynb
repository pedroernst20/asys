## ***Marco teorico usado***
Para transformar un numero complejo entre sus distintas representaciones (coordenada, binomial, polar y exponencial), es necesario conocer las relaciones fundamentales entre ellas.

Un n√∫mero complejo en el plano puede representarse como un par ordenado
(ùë•, ùë¶), donde ùë• es la parte real y ùë¶ la parte imaginaria. Esta representaci√≥n es equivalente a su forma **binomial**
$$
z = x +jy
$$

Para expresar un numero en su forma **polar** , se debe calcular el modulo *'r'* y su argumento *'Œ∏'* del mismo numero complejo, el modulo se obtiene aplicando el conocido teorema de pitagoras
$$
r = |z| = \sqrt{x^2+y^2}
$$
Y para obtener el argumento se obtiene con la funcion arcotangente
$$
\theta = arg(z) = arctan (\dfrac{y}{x})
$$
Entonces recopilando todo un numero en su representacion **polar** se podria escribir como
$$
z = (r, \theta)
$$

Ahora para expresar su forma **exponencial** usamos la formula de euler, se expresa como
$$
z = r e^{j\theta}
$$

Ejercicio 2.2 demostrar que el mapeo propuesto es una funci√≥n anal√≠tica en todo el plano complejo, empleando para ello dos m√©todos distintos:
* **M√©todo 1:** Usando la definici√≥n de los l√≠mites incrementales, esto es
<div align="center">
$$
\lim_{\Delta z \to 0} \frac{\Delta œâ}{\Delta z}
$$
</div>
<u> Resoluci√≥n: </u>
$$
\lim_{\Delta z \to 0} \frac{\Delta \omega}{\Delta z} = \lim_{\Delta z \to 0}\frac{f_{(z + \Delta z)} - f_{z}}{\Delta z}
$$ <br>


Se procede a expandir el numerador:
$$ f(z + \Delta z) - f(z) = (z + \Delta z)^2 - z^2 $$

$$ \require{cancel} f(z + \Delta z) - f(z) =  \cancel{ z^2 } + 2z \Delta z + \Delta z^2 -\cancel{ z^2 } $$

 <mark>  $$ f(z + \Delta z) - f(z) = 2z\Delta z + \Delta z^2 $$ </mark>
Volviendo al l√≠mite:
$$
\lim_{\Delta z \to 0} \frac{2z\Delta z + \Delta z^2}{\Delta z}
$$
$$
= \require{cancel} \lim_{\Delta z \to 0} \frac{\cancel{\Delta z}(2z + \Delta z)}{\cancel{\Delta z}}
$$
$$ = \lim_{\Delta z \to 0} 2z + \Delta z $$  <div align="center"><mark> ‚áí 2z </mark></div>
Se puede concluir que, al existir el l√≠mite y no depende de $\Delta z$, para todo $œâ = z^2$ es anal√≠tica en todo el plano $\mathbb{C}$.
* **M√©todo 2:** Verificar que las funciones $u_{(x;y)}$ y $v_{(x;y)}$ cumples las ecuaciones de Cauchy-Riemann en todo el plano.

<u> Resoluci√≥n: </u> <br>
Ecuaciones de Cauchy-Riemann (condici√≥n de analiticidad):
$$
\frac{‚àÇu}{\partial x} = \frac{‚àÇv}{\partial y} \land \frac{‚àÇu}{\partial y} = - \frac{‚àÇv}{\partial x}
$$ <br>
* $U_{(x;y)} = x^2 - y^2:$
$$
\frac{‚àÇu}{\partial x} = 2x
$$ <br>
$$
\frac{‚àÇu}{\partial y} = -2y
$$

* $V_{(x;y)} =  2xy:$
$$
\frac{‚àÇy}{\partial x} = 2y
$$ <br>
$$
\frac{‚àÇy}{\partial y} = 2x
$$

Entonces:

$$
\frac{‚àÇu}{\partial x} = \frac{‚àÇv}{\partial y}
$$ <br>
$$
2x = 2x \checkmark
$$ <br>
$$
\frac{‚àÇu}{\partial y} = \frac{‚àÇv}{\partial x}
$$ <br>
$$
-2y = 2y\checkmark
$$ <br>
Como cumple las ecuaciones de Cauchy-Riemann, la $\omega = z^2$ es anal√≠tica en todo el plano $\mathbb{C}$.
