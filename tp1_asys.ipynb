# ***Marco teorico usado***
Para transformar un numero complejo entre sus distintas representaciones (coordenada, binomial, polar y exponencial), es necesario conocer las relaciones fundamentales entre ellas.

Un n√∫mero complejo en el plano puede representarse como un par ordenado
(ùë•, ùë¶), donde ùë• es la parte real y ùë¶ la parte imaginaria. Esta representaci√≥n es equivalente a su forma **binomial**
$$
z = x +jy
$$

Para expresar un numero en su forma **polar** , se debe calcular el modulo *'r'* y su argumento *'Œ∏'* del mismo numero complejo, el modulo se obtiene aplicando el conocido teorema de pitagoras
$$
r = |z| = \sqrt{x^2+y^2}
$$
Y para obtener el argumento se obtiene con la funcion arcotangente:
$$
\theta = arg(z) = arctan (\dfrac{y}{x})
$$
Entonces recopilando todo un numero en su representacion **polar** se podria escribir como:
$$
z = (r, \theta)
$$

Ahora para expresar su forma **exponencial** usamos la formula de euler, se expresa como:
$$
z = r e^{j\theta}
$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "n": np.arange(1, 11),
    "modulo": [np.sqrt(13), 2, np.sqrt(5), np.sqrt(29), 8, 2, 2.14, 1, np.pi, 10]
}

df = pd.DataFrame(data)

df.plot(x="n", y="modulo", kind="line", marker="o", linestyle="-", color="b", legend=False)

plt.xlabel("n")
plt.ylabel("|z|")
plt.title("Gr√°fico discreto de los m√≥dulos de z (|z|)")
plt.xticks(df["n"])
plt.yticks(np.arange(0, 12, 1))
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "n": np.arange(1, 11),
    "argumento": [9*np.pi/50, np.pi/4, -np.pi/7, 1.87*np.pi, -np.pi/6, -np.pi/4, 0.88*np.pi, -np.pi/3, -2, np.pi/3]
}

df = pd.DataFrame(data)

df.plot(x="n", y="argumento", kind="line", marker="o", linestyle="-", color="b", legend=False)

plt.xlabel("n")
plt.ylabel(r"$\theta$ (radianes)")
plt.title(r"Gr√°fico discreto de los argumentos de z ($\theta$)")
plt.xticks(df["n"])
plt.yticks(np.arange(-3, 8, 1))
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
